````python
#question 1
#prelecture HW
import pandas as pd

# Load the dataset
url = 'https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-counties.csv'
lending_data = pd.read_csv(url)

# Check for missing values
missing_values = lending_data.isnull().sum()

# Display the missing value count
print(missing_values)
````

````python
#question 3
# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(lending_data.head())

# Summary of each column
print("\nSummary of the dataset columns:")

# For categorical columns like 'county' and 'state'
print("\n'county' column:")
print(lending_data['county'].describe())

print("\n'state' column:")
print(lending_data['state'].describe())

# For numerical columns like 'cases' and 'deaths'
print("\n'cases' column:")
print(lending_data['cases'].describe())

print("\n'deaths' column:")
print(lending_data['deaths'].describe())

# For the 'date' column
print("\n'date' column:")
print(lending_data['date'].describe())

# Additional summary: Data types and missing values
print("\nData types and missing values:")
print(lending_data.info())

lending_data.describe()
lending_data.value_counts()
````

````python
#question 4
missing_values = lending_data.isnull().sum()

print("\nMissing values in each column:")
print(missing_values[missing_values > 0])

# Additional: percentage of missing values
missing_percentage = (missing_values / len(lending_data)) * 100
print("\nPercentage of missing values in each column:")
print(missing_percentage[missing_percentage > 0])
````

````python
non_numeric_cols = lending_data.select_dtypes(exclude=['number']).columns

print("Non-numeric columns and their unique values:")
for col in non_numeric_cols:
    unique_values = lending_data[col].unique()
    print(f"\nColumn: {col}")
    print(f"Unique values (sample): {unique_values[:10]}")  # Show a sample of unique values
    print(f"Number of unique values: {len(unique_values)}")
lending_data.shape
lending_data.describe()
````

df.shape: This method gives the dimensions of the DataFrame, which includes the number of rows and columns. For this dataset, calling df.shape would give you something like (n_rows, n_columns).

2. Values Reported in the "count" Column
df.shape: This provides the total number of rows in the dataset. If the dataset has missing values in some columns, the shape will still reflect the total number of rows, regardless of how many valid data points are present in each column.

df.describe(): The "count" value in df.describe() shows the number of non-null entries for each numeric column. If a column has missing values, the count will be less than the total number of rows, reflecting only the number of non-null entries.

###In Python, especially in the context of libraries like pandas, attributes and methods serve different purposes: ### Attributes - **Definition**: Attributes are values or properties associated with an object. They do not require parentheses to access. - **Usage**: Attributes are used to get information about the object or to store data. - **Examples**: - `df.shape`: Returns a tuple representing the dimensions of the DataFrame (number of rows and columns). No parentheses are needed because `shape` is an attribute. - `df.columns`: Returns the column labels of the DataFrame. Again, no parentheses are needed because `columns` is an attribute. ```python print(df.shape) # (number of rows, number of columns) print(df.columns) # Index of column names ``` ### Methods - **Definition**: Methods are functions associated with an object. They perform actions or computations and require parentheses to call. - **Usage**: Methods are used to perform operations, computations, or transformations on the object. - **Examples**: - `df.describe()`: Generates descriptive statistics of the DataFrame. Parentheses are needed because `describe` is a method that performs an operation and returns the result. - `df.head()`: Returns the first few rows of the DataFrame. Parentheses are used to call the method and optionally specify the number of rows to return.
