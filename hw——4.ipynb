To create a 95% confidence interval that covers approximately 95% of the bootstrapped sample means using the standard error of the mean (SEM), you can follow these steps:

Calculate the Standard Error of the Mean (SEM):

For each bootstrap sample, compute the sample mean.
Calculate the standard deviation of these bootstrapped sample means.
The SEM is essentially this standard deviation.
Determine the Critical Value:

For a 95% confidence interval and assuming a normal distribution, the critical z-value is approximately 1.96.
If your sample size is small and the population standard deviation is unknown, you might use the t-distribution instead.
Compute the Confidence Interval:

Lower Bound = Sample Mean − (Critical Value × SEM)
Upper Bound = Sample Mean + (Critical Value × SEM)
So, the 95% confidence interval is: Sample Mean ± 1.96 × SEM

```python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import numpy as np

# Load the penguins dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv"
penguins = pd.read_csv(url)
penguins = penguins.dropna()
penguins
```

```python
sample = penguins['flipper_length_mm']


def bootstrap_ci(sample, n_bootstrap=1000, statistic=np.mean, alpha=0.05):
   
    bootstrap_samples = np.random.choice(sample, size=(n_bootstrap, len(sample)), replace=True)
    bootstrap_statistics = np.apply_along_axis(statistic, 1, bootstrap_samples)
    lower_bound = np.percentile(bootstrap_statistics, 100 * alpha / 2)
    upper_bound = np.percentile(bootstrap_statistics, 100 * (1 - alpha / 2))
    return lower_bound, upper_bound


# Compute a 95% confidence interval for the population mean
ci_mean = bootstrap_ci(sample, statistic=np.mean)
print(f"95% Bootstrap Confidence Interval for the Mean: {ci_mean}")
# Compute a 95% confidence interval for the population median
ci_median = bootstrap_ci(sample, statistic=np.median)
print(f"95% Bootstrap Confidence Interval for the Median: {ci_median}")
```

We need to distinguish between the population parameter and the sample statistic when dealing with confidence intervals because:

Population Parameter: This is the true value we're trying to estimate (like the actual average of a whole population). It's unknown and fixed.

Sample Statistic: This is what we calculate from the sample data we collect (like the average of the sample). It's known because we measured it.

The confidence interval uses the sample statistic to give us a range of possible values that might contain the true population parameter. We make this distinction because the population parameter is what we're interested in, but we can only estimate it using the sample statistic, which is based on limited data.

&&The concept of a confidence interval is specifically for **population parameters**, not sample parameters. Here's how it works: 1. **Population Parameter**: - This is the unknown value we want to estimate (e.g., population mean, population proportion). - The confidence interval is constructed to estimate this unknown population parameter. 2. **Sample Statistic**: - This is the value we calculate from the sample data (e.g., sample mean, sample proportion). - The confidence interval is built around this sample statistic to provide a range of likely values for the population parameter. ### Summary: The confidence interval is a range that estimates the **unknown population parameter**, using the **sample statistic** as the basis for that estimate.

```python
#bootstrapping
np.random.seed(42)


def bootstrap_mean_diff(data, n_iterations=1000):
    bootstrapped_diffs = []
    n = len(data)
    
    
    for _ in range(n_iterations):
        resampled_data = data.sample(n, replace=True)
        mean_diff = resampled_data['FinalHealthScore'].mean() - resampled_data['InitialHealthScore'].mean()
        bootstrapped_diffs.append(mean_diff)
    
    return np.array(bootstrapped_diffs)


n_iterations = 10000
bootstrapped_diffs = bootstrap_mean_diff(df, n_iterations)


observed_diff = df['FinalHealthScore'].mean() - df['InitialHealthScore'].mean()

#checking p-values
p_value = np.sum(bootstrapped_diffs >= observed_diff) / n_iterations

print(f"Observed Mean Difference: {observed_diff}")
print(f"P-value from bootstrapping: {p_value}")
```

```python
scores = df['InitialHealthScore']


np.random.seed(42)

#bootstrapping
def bootstrap(data, n_bootstrap=1000, confidence_level=0.95):
   
    boot_means = np.empty(n_bootstrap)
    
   
    for i in range(n_bootstrap):
        boot_sample = np.random.choice(data, size=len(data), replace=True)
        boot_means[i] = np.mean(boot_sample)
    
#confidence interval  
    alpha = 1 - confidence_level
    lower_bound = np.percentile(boot_means, 100 * (alpha / 2))
    upper_bound = np.percentile(boot_means, 100 * (1 - alpha / 2))
    
    return np.mean(boot_means), (lower_bound, upper_bound)


mean_score, confidence_interval = bootstrap(scores)

print(f"Bootstrap Mean: {mean_score}")
print(f"95% Confidence Interval: {confidence_interval}")
```
