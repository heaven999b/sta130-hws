```python
import pandas as pd
cols = pd.read_csv("https://raw.githubusercontent.com/pointOfive/stat130chat130/main/CP/var_names.csv")
data = pd.read_csv("https://raw.githubusercontent.com/pointOfive/stat130chat130/main/CP/CSCS_data_anon.csv",
                   na_values=["9999", "", " ", "Presented but no response", "NA"])
empty = (data.isna().sum()==data.shape[0])
data = data[empty.index[~empty]] # keep non empty columns only

data.shape
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 删除缺失值以确保可视化正常
data_cleaned = data['CONNECTION_social_num_close_friends'].dropna()

# 直方图
plt.figure(figsize=(8, 6))
plt.hist(data_cleaned, bins=20, edgecolor='black', alpha=0.7)
plt.title('Distribution of CONNECTION_social_num_close_friends (Histogram)')
plt.xlabel('Number of Close Friends')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd

# 假设 data 是你的 DataFrame
# 创建新列，将数字分类为 'lots' 或 'small'
data['friend_category'] = data['CONNECTION_social_num_close_friends'].apply(
    lambda x: 'lots' if x > 8 else 'small'
)

# 查看结果
print(data[['CONNECTION_social_num_close_friends', 'friend_category']])
```

```python
import matplotlib.pyplot as plt

# 删除缺失值
data_cleaned = data['LONELY_direct'].dropna()

# 绘制直方图
plt.figure(figsize=(8, 6))
plt.hist(data_cleaned, bins=20, edgecolor='black', alpha=0.7)
plt.title('Distribution of LONELY_direct')
plt.xlabel('LONELY_direct')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 提取 LONELY_direct 列中的唯一值
unique_values = data['LONELY_direct'].dropna().unique()

# 查看种类
print("LONELY_direct 列中的种类:", unique_values)
```

```python
# 创建新列，将符合第一个类别的标记为 high_lonely，其余的标记为 low_lonely
data['loneliness_level'] = data['LONELY_direct'].apply(
    lambda x: 'high_lonely' if x == 'Occasionally or a moderate amount of time (e.g. 3-4 days)' else 'low_lonely'
)

# 展示新列的前五项
print(data[['LONELY_direct', 'loneliness_level']].head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.preprocessing import LabelEncoder

# 准备数据：编码分类变量
le_loneliness = LabelEncoder()
le_friend = LabelEncoder()

new_data['loneliness_level_encoded'] = le_loneliness.fit_transform(new_data['loneliness_level'])
new_data['friend_category_encoded'] = le_friend.fit_transform(new_data['friend_category'])

# 特征和目标
X = new_data[['friend_category_encoded']]
y = new_data['loneliness_level_encoded']

# 划分训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 创建和训练决策树模型
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)

# 测试模型
y_pred = clf.predict(X_test)

# 绘制混淆矩阵
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=le_loneliness.classes_)
disp.plot(cmap='Blues')
plt.title('Confusion Matrix')
plt.show()

# 可视化决策树
plt.figure(figsize=(12, 8))
plot_tree(clf, feature_names=['friend_category_encoded'], class_names=le_loneliness.classes_, filled=True)
plt.title('Decision Tree')
plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, ConfusionMatrixDisplay

# 计算混淆矩阵
cm = confusion_matrix(y_test, y_pred)

# 从混淆矩阵提取 TN, FP, FN, TP
tn, fp, fn, tp = cm.ravel()

# 计算性能指标
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, pos_label=1)  # 针对 high_lonely 的精确度
recall = recall_score(y_test, y_pred, pos_label=1)        # 针对 high_lonely 的召回率（灵敏度）
specificity = tn / (tn + fp)                             # 特异性
f1 = f1_score(y_test, y_pred, pos_label=1)               # F1 分数

# 打印结果
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall (Sensitivity): {recall:.2f}")
print(f"Specificity: {specificity:.2f}")
print(f"F1 Score: {f1:.2f}")

# 可视化混淆矩阵
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=le_loneliness.classes_)
disp.plot(cmap='Blues')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc
y_proba = clf.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_proba)
auc_score = auc(fpr, tpr)

plt.plot(fpr, tpr, label=f'AUC = {auc_score:.2f}')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```
